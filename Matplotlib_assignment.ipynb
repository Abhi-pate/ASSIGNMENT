### QUESTION-2

```python
# A scatter plot is a diagram where each value in the data set is represented by a dot


import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.xlabel("x_axis")
plt.ylabel("y_axis")
plt.title("Scatter_Plot")
plt.scatter(x,y)
plt.show()
```

### QUESTION-3

```python
# subplot( m , n , p ) divides the current figure into an m -by- n grid and creates axes in the position specified by p


import numpy as np
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2,2,1)
plt.plot(x,y)
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2,2,2)
plt.plot(x,y)
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2,2,3)
plt.plot(x,y)
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2,2,4)
plt.plot(x,y)
```

### QUESTION-4

```python
# Bar plots are a type of data visualization used to represent data in the form of rectangular bars. mostly used when one axis is string and another is numerical value.

import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.subplot(2,2,1)
plt.xlabel("Company_Name")
plt.ylabel("Profit")
plt.bar(company, profit)
plt.subplot(2,2,4)
plt.xlabel("profit")
plt.ylabel("company_name")
plt.barh(company, profit)
plt.show()
plt.show()
```

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
plt.subplot(1,2,1)
plt.xlabel("Box1")
plt.boxplot(box1)
plt.subplot(1,2,2)
plt.xlabel("Box2")
plt.boxplot(box2)
plt.show()
```
